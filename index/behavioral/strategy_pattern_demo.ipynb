```python
# å¯¼å…¥æ‰€éœ€æ¨¡å—
import sys
sys.path.append('../..')

from utils.client import AIClient, MockAIClient
from abc import ABC, abstractmethod
from typing import Dict, Any, List
import time
import random

print("ğŸ¯ Strategy Pattern Demo - AI Provider Selection")
print("=" * 50)
print("Strategy Pattern = ä¸€ä¸ªæ¥å£ï¼Œå¤šç§å®ç°ï¼ŒåŠ¨æ€é€‰æ‹©æœ€ä½³ç­–ç•¥ï¼")
```

## ğŸ“‹ Step 1: å®šä¹‰ç­–ç•¥æ¥å£

å…ˆå®šä¹‰ç»Ÿä¸€çš„"AIå¤„ç†ç­–ç•¥"æ¥å£ï¼Œè¿™æ ·æ‰€æœ‰å…·ä½“ç­–ç•¥éƒ½è¦å®ç°ç›¸åŒçš„æ–¹æ³•ï¼š

```python
class AIProcessingStrategy(ABC):
    """AIå¤„ç†ç­–ç•¥çš„æŠ½è±¡åŸºç±» - å®šä¹‰ç»Ÿä¸€æ¥å£"""
    
    @abstractmethod
    def process_query(self, query: str, context: Dict[str, Any]) -> Dict[str, Any]:
        """å¤„ç†æŸ¥è¯¢çš„æ ¸å¿ƒæ–¹æ³•"""
        pass
    
    @abstractmethod
    def get_strategy_info(self) -> Dict[str, Any]:
        """è·å–ç­–ç•¥ä¿¡æ¯"""
        pass
    
    @abstractmethod
    def estimate_cost(self, query: str) -> float:
        """ä¼°ç®—å¤„ç†æˆæœ¬"""
        pass

print("âœ… ç­–ç•¥æ¥å£å®šä¹‰å®Œæˆ!")
print("ğŸ“ æ‰€æœ‰å…·ä½“ç­–ç•¥éƒ½å¿…é¡»å®ç°: process_query(), get_strategy_info(), estimate_cost()")
```

## ğŸ­ Step 2: å®ç°å…·ä½“çš„AIç­–ç•¥

ç°åœ¨åˆ›å»ºä¸åŒçš„AIå¤„ç†ç­–ç•¥ï¼Œæ¯ä¸ªéƒ½æœ‰è‡ªå·±çš„ç‰¹ç‚¹ï¼š

```python
class FastModelStrategy(AIProcessingStrategy):
    """å¿«é€Ÿæ¨¡å‹ç­–ç•¥ - è¿½æ±‚é€Ÿåº¦ï¼Œé€‚åˆç®€å•ä»»åŠ¡"""
    
    def __init__(self):
        self.client = MockAIClient("fast-model")
        self.name = "FastModel"
    
    def process_query(self, query: str, context: Dict[str, Any]) -> Dict[str, Any]:
        start_time = time.time()
        
        # æ¨¡æ‹Ÿå¿«é€Ÿå¤„ç†
        time.sleep(0.1)  # å¾ˆå¿«ï¼
        response = self.client.simple_query(f"Quick answer for: {query}")
        
        process_time = time.time() - start_time
        
        return {
            "strategy": self.name,
            "response": response,
            "processing_time": round(process_time, 2),
            "cost": self.estimate_cost(query),
            "quality_score": 6.5  # è´¨é‡ä¸€èˆ¬ä½†å¾ˆå¿«
        }
    
    def get_strategy_info(self) -> Dict[str, Any]:
        return {
            "name": self.name,
            "speed": "âš¡ æå¿«",
            "cost": "ğŸ’° ä¾¿å®œ",
            "quality": "ğŸ“ ä¸€èˆ¬",
            "best_for": "ç®€å•é—®ç­”ã€å¿«é€Ÿå“åº”éœ€æ±‚"
        }
    
    def estimate_cost(self, query: str) -> float:
        return len(query) * 0.001  # å¾ˆä¾¿å®œ


class PremiumModelStrategy(AIProcessingStrategy):
    """é«˜çº§æ¨¡å‹ç­–ç•¥ - è¿½æ±‚è´¨é‡ï¼Œé€‚åˆå¤æ‚ä»»åŠ¡"""
    
    def __init__(self):
        self.client = MockAIClient("premium-model")
        self.name = "PremiumModel"
    
    def process_query(self, query: str, context: Dict[str, Any]) -> Dict[str, Any]:
        start_time = time.time()
        
        # æ¨¡æ‹Ÿé«˜è´¨é‡å¤„ç†ï¼Œéœ€è¦æ›´å¤šæ—¶é—´
        time.sleep(0.5)  # æ…¢ä¸€äº›ï¼Œä½†è´¨é‡é«˜
        response = self.client.simple_query(f"Detailed analysis for: {query}")
        
        process_time = time.time() - start_time
        
        return {
            "strategy": self.name,
            "response": response + " [Detailed reasoning and examples provided]",
            "processing_time": round(process_time, 2),
            "cost": self.estimate_cost(query),
            "quality_score": 9.2  # é«˜è´¨é‡
        }
    
    def get_strategy_info(self) -> Dict[str, Any]:
        return {
            "name": self.name,
            "speed": "ğŸŒ è¾ƒæ…¢",
            "cost": "ğŸ’ æ˜‚è´µ",
            "quality": "ğŸ† æé«˜",
            "best_for": "å¤æ‚åˆ†æã€åˆ›æ„å†™ä½œã€æ·±åº¦æ€è€ƒ"
        }
    
    def estimate_cost(self, query: str) -> float:
        return len(query) * 0.01  # è¾ƒè´µ


class BalancedModelStrategy(AIProcessingStrategy):
    """å¹³è¡¡æ¨¡å‹ç­–ç•¥ - é€Ÿåº¦ä¸è´¨é‡çš„å¹³è¡¡"""
    
    def __init__(self):
        self.client = MockAIClient("balanced-model")
        self.name = "BalancedModel"
    
    def process_query(self, query: str, context: Dict[str, Any]) -> Dict[str, Any]:
        start_time = time.time()
        
        # æ¨¡æ‹Ÿå¹³è¡¡å¤„ç†
        time.sleep(0.3)  # ä¸­ç­‰é€Ÿåº¦
        response = self.client.simple_query(f"Balanced response for: {query}")
        
        process_time = time.time() - start_time
        
        return {
            "strategy": self.name,
            "response": response,
            "processing_time": round(process_time, 2),
            "cost": self.estimate_cost(query),
            "quality_score": 8.0  # å¹³è¡¡çš„è´¨é‡
        }
    
    def get_strategy_info(self) -> Dict[str, Any]:
        return {
            "name": self.name,
            "speed": "âš–ï¸ ä¸­ç­‰",
            "cost": "ğŸ’µ ä¸­ç­‰",
            "quality": "ğŸ‘ è‰¯å¥½",
            "best_for": "å¤§éƒ¨åˆ†æ—¥å¸¸ä»»åŠ¡çš„æœ€ä½³é€‰æ‹©"
        }
    
    def estimate_cost(self, query: str) -> float:
        return len(query) * 0.005  # ä¸­ç­‰ä»·æ ¼

print("âœ… ä¸‰ç§AIç­–ç•¥å®ç°å®Œæˆ!")
print("ğŸš€ FastModel: å¿«é€Ÿä¾¿å®œ")
print("ğŸ’ PremiumModel: é«˜è´¨é‡æ˜‚è´µ")
print("âš–ï¸ BalancedModel: å¹³è¡¡é€‰æ‹©")
```

## ğŸ›ï¸ Step 3: æ™ºèƒ½ç­–ç•¥é€‰æ‹©å™¨

è¿™æ˜¯Strategy Patternçš„æ ¸å¿ƒ - æ ¹æ®ä¸åŒæƒ…å†µé€‰æ‹©æœ€ä½³ç­–ç•¥ï¼š

```python
class SmartAIProcessor:
    """æ™ºèƒ½AIå¤„ç†å™¨ - Strategy Patternçš„ä¸Šä¸‹æ–‡ç±»"""
    
    def __init__(self):
        # åˆå§‹åŒ–æ‰€æœ‰å¯ç”¨ç­–ç•¥
        self.strategies = {
            "fast": FastModelStrategy(),
            "premium": PremiumModelStrategy(), 
            "balanced": BalancedModelStrategy()
        }
        
        # é»˜è®¤ç­–ç•¥
        self.current_strategy = self.strategies["balanced"]
        self.processing_history = []
    
    def set_strategy(self, strategy_name: str):
        """æ‰‹åŠ¨è®¾ç½®ç­–ç•¥"""
        if strategy_name in self.strategies:
            self.current_strategy = self.strategies[strategy_name]
            print(f"ğŸ”„ ç­–ç•¥åˆ‡æ¢åˆ°: {strategy_name}")
        else:
            print(f"âŒ æœªçŸ¥ç­–ç•¥: {strategy_name}")
    
    def auto_select_strategy(self, query: str, context: Dict[str, Any]) -> str:
        """æ ¹æ®æŸ¥è¯¢å’Œä¸Šä¸‹æ–‡è‡ªåŠ¨é€‰æ‹©æœ€ä½³ç­–ç•¥"""
        query_lower = query.lower()
        
        # æ™ºèƒ½å†³ç­–é€»è¾‘
        if context.get("priority") == "speed":
            return "fast"
        elif context.get("priority") == "quality":
            return "premium"
        elif context.get("budget_limited", False):
            return "fast"
        elif any(word in query_lower for word in ["analyze", "complex", "detailed", "creative", "write"]):
            return "premium"
        elif any(word in query_lower for word in ["quick", "simple", "what is", "define"]):
            return "fast"
        else:
            return "balanced"
    
    def process(self, query: str, context: Dict[str, Any] = None, auto_select: bool = True) -> Dict[str, Any]:
        """å¤„ç†æŸ¥è¯¢ - Strategy Patternçš„æ ¸å¿ƒæ–¹æ³•"""
        if context is None:
            context = {}
        
        # è‡ªåŠ¨é€‰æ‹©ç­–ç•¥ï¼ˆå¦‚æœå¼€å¯ï¼‰
        if auto_select:
            best_strategy_name = self.auto_select_strategy(query, context)
            self.current_strategy = self.strategies[best_strategy_name]
            print(f"ğŸ¯ è‡ªåŠ¨é€‰æ‹©ç­–ç•¥: {best_strategy_name}")
        
        # ä½¿ç”¨å½“å‰ç­–ç•¥å¤„ç†æŸ¥è¯¢
        result = self.current_strategy.process_query(query, context)
        
        # è®°å½•å¤„ç†å†å²
        self.processing_history.append({
            "query": query[:50] + "...",
            "strategy_used": result["strategy"],
            "processing_time": result["processing_time"],
            "cost": result["cost"]
        })
        
        return result
    
    def compare_strategies(self, query: str, context: Dict[str, Any] = None) -> Dict[str, Any]:
        """æ¯”è¾ƒæ‰€æœ‰ç­–ç•¥çš„è¡¨ç°"""
        if context is None:
            context = {}
            
        print(f"ğŸ”¬ æ¯”è¾ƒæ‰€æœ‰ç­–ç•¥å¤„ç†æŸ¥è¯¢: {query[:50]}...\n")
        
        results = {}
        for name, strategy in self.strategies.items():
            print(f"æµ‹è¯•ç­–ç•¥: {name}")
            result = strategy.process_query(query, context)
            results[name] = result
            
            print(f"  â±ï¸  å¤„ç†æ—¶é—´: {result['processing_time']}ç§’")
            print(f"  ğŸ’° æˆæœ¬: ${result['cost']:.4f}")
            print(f"  â­ è´¨é‡åˆ†: {result['quality_score']}/10")
            print()
        
        return results
    
    def get_available_strategies(self) -> Dict[str, Dict[str, Any]]:
        """è·å–æ‰€æœ‰å¯ç”¨ç­–ç•¥çš„ä¿¡æ¯"""
        return {name: strategy.get_strategy_info() 
                for name, strategy in self.strategies.items()}
    
    def get_processing_stats(self) -> Dict[str, Any]:
        """è·å–å¤„ç†ç»Ÿè®¡ä¿¡æ¯"""
        if not self.processing_history:
            return {"message": "è¿˜æ²¡æœ‰å¤„ç†è®°å½•"}
        
        total_cost = sum(record["cost"] for record in self.processing_history)
        total_time = sum(record["processing_time"] for record in self.processing_history)
        strategy_usage = {}
        
        for record in self.processing_history:
            strategy = record["strategy_used"]
            strategy_usage[strategy] = strategy_usage.get(strategy, 0) + 1
        
        return {
            "total_requests": len(self.processing_history),
            "total_cost": round(total_cost, 4),
            "total_time": round(total_time, 2),
            "strategy_usage": strategy_usage,
            "most_used_strategy": max(strategy_usage.keys(), key=lambda k: strategy_usage[k])
        }

print("âœ… SmartAIProcessor åˆ›å»ºå®Œæˆ!")
print("ğŸ§  å…·å¤‡æ™ºèƒ½ç­–ç•¥é€‰æ‹©èƒ½åŠ›")
print("ğŸ“Š æ”¯æŒç­–ç•¥æ¯”è¾ƒå’Œç»Ÿè®¡åˆ†æ")
```

## ğŸš€ Step 8: å®é™…åº”ç”¨åœºæ™¯ - çœŸå®ä¸–ç•Œçš„Strategy Pattern

è®©æˆ‘ä»¬çœ‹çœ‹Strategy Patternåœ¨å®é™…AIç³»ç»Ÿä¸­çš„åº”ç”¨ä»·å€¼ï¼š

```python
print("\n" + "=" * 60)
print("ğŸŒ å®é™…åº”ç”¨åœºæ™¯æ¼”ç¤º")
print("=" * 60)

# æ¨¡æ‹ŸçœŸå®ä¸šåŠ¡åœºæ™¯
real_world_scenarios = [
    {
        "scenario": "ç”µå•†å®¢æœç³»ç»Ÿ",
        "context": {"user_type": "vip", "priority": "quality"},
        "query": "I have a problem with my order delivery",
        "expected_strategy": "premium"
    },
    {
        "scenario": "åœ¨çº¿æ•™è‚²å¹³å°",
        "context": {"student_level": "basic"},
        "query": "What is photosynthesis?",
        "expected_strategy": "fast"
    },
    {
        "scenario": "ä¼ä¸šå’¨è¯¢æœåŠ¡",
        "context": {"client_tier": "enterprise"},
        "query": "Analyze our market positioning strategy",
        "expected_strategy": "premium"
    },
    {
        "scenario": "ç§»åŠ¨AppåŠ©æ‰‹",
        "context": {"device": "mobile", "network": "3g"},
        "query": "Set a reminder for 3pm",
        "expected_strategy": "fast"
    }
]

print("ğŸ¬ æ¨¡æ‹ŸçœŸå®ä¸šåŠ¡åœºæ™¯...\n")

for scenario in real_world_scenarios:
    print(f"ğŸ¢ åœºæ™¯: {scenario['scenario']}")
    print(f"ğŸ“ ç”¨æˆ·æŸ¥è¯¢: {scenario['query']}")
    print(f"âš™ï¸ ä¸šåŠ¡ä¸Šä¸‹æ–‡: {scenario['context']}")
    
    # åŸºäºä¸šåŠ¡ä¸Šä¸‹æ–‡çš„æ™ºèƒ½ç­–ç•¥é€‰æ‹©
    if scenario['context'].get('user_type') == 'vip' or scenario['context'].get('client_tier') == 'enterprise':
        chosen_strategy = 'premium'
    elif scenario['context'].get('device') == 'mobile' or scenario['context'].get('student_level') == 'basic':
        chosen_strategy = 'fast'
    else:
        chosen_strategy = 'balanced'
    
    ai_processor.set_strategy(chosen_strategy)
    result = ai_processor.process(scenario['query'], scenario['context'], auto_select=False)
    
    print(f"âœ… é€‰æ‹©ç­–ç•¥: {result['strategy']}")
    print(f"ğŸ’¡ ä¸šåŠ¡ä»·å€¼: {'é«˜ç«¯å®¢æˆ·ä¼˜è´¨æœåŠ¡' if chosen_strategy == 'premium' else 'å¿«é€Ÿå“åº”ç”¨æˆ·éœ€æ±‚' if chosen_strategy == 'fast' else 'å¹³è¡¡æˆæœ¬ä¸è´¨é‡'}")
    print("-" * 50)

print("\nğŸ¯ Strategy Patternåœ¨å®é™…ä¸šåŠ¡ä¸­çš„ä»·å€¼:")
print("   âœ¨ ç”¨æˆ·ä½“éªŒä¸ªæ€§åŒ–")
print("   âœ¨ èµ„æºé…ç½®ä¼˜åŒ–")
print("   âœ¨ æˆæœ¬æ§åˆ¶çµæ´»")
print("   âœ¨ ä¸šåŠ¡è§„åˆ™æ˜“æ‰©å±•")
```
